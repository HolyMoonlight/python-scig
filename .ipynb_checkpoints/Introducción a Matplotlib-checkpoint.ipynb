## Introducción

Matplotlib es una librearía de Python para graficación que se integra a `numpy` y es utilizada por la mayoría de librerías como forma de reportar graficas.

Con `matplotlib` podemos crear gráficas en 2 y 3 dimensiones. La documentación puede consultarse en https://matplotlib.org/gallery/index.html.

Una de las gráficas más importantes son las series las cuales nos permiten graficar un conjunto de puntos para dos ejes.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)

plt.show()
```

## Sistema marcadores

Matplotlib posee un conjunto nombrado de marcadores disponible en https://matplotlib.org/api/markers_api.html.

Los principales marcadores son:

`-` - línea

`--` - línea punteada

`.` - punto

`o` - punto grande

`^` - triángulo hacia arriba

`s` - cuadrado

`p` - pentágono

`h` - hexágono

`*` - estrella

`+` - signo más

`x` - signo tache

Los marcadores básicos se pueden combinar con la línea y la línea punteada, por ejemplo, `s-`, `*--`.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 1, 6]

plt.plot(x, y, "s--")

plt.show()
```

## Sistema de colores

Algunos colores importantes podemos encontrarlos en https://www.w3schools.com/tags/ref_colornames.asp.

Los colores más útiles son:

`tomato` - rojo tomate

`deeppink` - rosa intenso

`slateblue` - azul marino

`green` - verde

`yellowgreen` - verde amarillo

`lime` - verde lima

`yellow` - amarillo

También podemos usar colores hexadecimales como `#F855DD`.

```python
x = [1, 2, 3, 4, 5]
y = [1, 0, 2, 1, 0]

plt.plot(x, y, color="red")

plt.show()
```

## Graficar varias series

Para graficar varias series podemos utilizar `plt.plot(...)` varias veces (una por serie) antes de hacer `plt.show()` o `plt.savefig("g.png")`.

```python
x1 = [1, 2, 3, 4, 5]
y1 = [5, 4, 2, 7, 6]

plt.plot(x1, y1, "*-", color="blue")

x2 = [1, 2, 3]
y2 = [6, 3, 4]

plt.plot(x2, y2, "+--", color="red")

x3 = [1, 2, 3, 4, 5, 6]
y3 = [1, 2, 5, 4, 8, 7]

plt.plot(x3, y3, "o", color="green")

plt.show()
```

## Integrar arreglos de `numpy` a `matplotlib`

Podemos usar los arreglos de numpy para crear graficas más sofisticadas y basadas en funciones matemáticas.

```python
import numpy as np

# Crear un espacio lineal (un arreglo de `n` valores entre `a` y `b` equidistantes)
x = np.linspace(-10, 10, 11)

print(x)
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)

plt.plot(x, y)

plt.show()
```

```python
# Generar la función seno perturbada aleatoriamente
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = y1 + (2 * np.random.random(100) - np.ones(100)) * 0.2

plt.plot(x, y1, color="red")
plt.plot(x, y2, ".", color="cyan")

plt.show()
```

## Graficas de pastel (pie)

Una gráfica muy utilizada en los reportes es la gráfica de pastel o pie. La cual representa un círuclo dividido en categorías dónde cada categoría equivale a un porcentaje del pastel depiendiendo el número de elementos (su peso) respecto al total de elementos.

```python
categorias = ["manzanas", "peras", "mangos"]
pesos = [15, 28, 6]

plt.pie(pesos)
plt.axis("equal")

plt.show()
```

```python
categorias = ["manzanas", "peras", "mangos"]
pesos = [15, 28, 6]

plt.pie(pesos, labels=categorias)
plt.axis("equal")

plt.show()
```

```python
categorias = ["manzanas", "peras", "mangos"]
pesos = [15, 28, 6]
colores = ["red", "yellowgreen", "gold"]

plt.pie(pesos, labels=categorias, colors=colores)
plt.axis("equal")

plt.show()
```

```python
categorias = ["manzanas", "peras", "mangos"]
pesos = [15, 28, 6]
colores = ["red", "yellowgreen", "gold"]

# Formato: %f - float, %.nf - float con n cifras, %% - signo %
plt.pie(pesos, labels=categorias, colors=colores, autopct="%.1f%%")
plt.axis("equal")

plt.show()
```

## Graficas de barras

Las gráficas de barras nos permiten representar información de un eje contra su altura o peso. Por ejemplo, son útiles para mostrar las diferencias entre una misma categoría vs una subcategoría, como las edades entre hombres, las diferencias de PH en un experimiento, entre otras.

Para definir las barras a mostrar, necesitamos especificar la posición y altura de cada barra en dos arreglos diferentes.

```python
posiciones = [1, 2, 3, 4, 5]
alturas = [10, 20, 5, 7, 23]

plt.bar(posiciones, alturas)

plt.show()
```

Adicionalmente podemos especificar como tercer parámetro el ancho de la barra. Esto es muy útil cuándo se quieren mostrar barras de distintas categorías en una misma gráfica.

```python
posiciones = [1, 2, 3, 4, 5]
alturas = [10, 20, 5, 7, 23]

plt.bar(posiciones, alturas, 0.1)

plt.show()
```

```python
edades_hombres = [35, 23, 17, 55, 89]
edades_mujeres = [30, 27, 16, 45, 57]

d = 0.2
posiciones_hombres = [1 - d, 2 - d, 3 - d, 4 - d, 5 - d] # np.arange(5) - d
posiciones_mujeres = np.arange(1, 6) + d

plt.bar(posiciones_hombres, edades_hombres, 2 * d, color="cornflowerblue")
plt.bar(posiciones_mujeres, edades_mujeres, 2 * d, color="hotpink")

plt.show()
```

```python
h = [35, 23, 17, 55, 89]
m = [30, 27, 16, 55, 87]

n = len(h)

w = 0.2

plt.bar(np.arange(n) - w, h, 2 * w, color="cornflowerblue")
plt.bar(np.arange(n) + w, m, 2 * w, color="hotpink")

plt.show()
```
